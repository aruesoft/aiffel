## 이미지 부르기 & 자르기 & 저장

```python
from PIL import Image, ImageColor
import os
```

```python
img_path = "logo.png"
img = Image.open(img_path)
print(img_path)
print(type(img))
img
```

```python
img.size
```

```python
# 이미지 사이즈 
W, H = img.size
print((W, H))

#이미지 포맷 정보 
print(img.format)
print(img.size)
print(img.mode)
```

```python
#이미지 자르기
img.crop((30,30,100,100))
```

```python
# 새로운 이미지 파일명
cropped_img_path = "cropped_img.png"
img.crop((30,30,100,100)).save(cropped_img_path)
print("저장 완료!")
```

```python
print(cropped_img_path)
```

```python
img_cropped = Image.open(cropped_img_path)
```
