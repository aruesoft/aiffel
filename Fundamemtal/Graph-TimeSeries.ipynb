```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
display(flights.shape)
display(flights.info())
```

```python
flights.describe()
```

```python
flights['month2'] = pd.to_datetime(flights['month'],format='%b').dt.month
```

```python
flights.info()
```

```python
flights['date'] = flights['year'].astype(str)+'-'+flights['month2'].astype(str)
```

```python
flights['date'] = pd.to_datetime(flights['date'])
```

```python
flights.info()
```

```python
sns.barplot(data=flights,x='date',y='passengers')
```

```python
sns.barplot(data=flights,x='year',y='passengers')
```

```python
sns.barplot(data=flights,x='year',y='passengers',hue='month2')
```

```python
sns.pointplot(data=flights,x='date',y='passengers')
```

```python
sns.lineplot(data=flights,x='date',y='passengers')
```

```python
# Q. seaborn pointplot을 그려봅시다.
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
sns.pointplot(data=flights, x='year', y='passengers')

ax2 = fig.add_subplot(2,1,2)
sns.pointplot(data=flights, x='date', y='passengers')
```

```python
# Q. seaborn lineplot을 그려봅시다.
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
sns.lineplot(data=flights, x='year', y='passengers')

ax2 = fig.add_subplot(2,1,2)
sns.lineplot(data=flights, x='date', y='passengers')
```

```python
sns.lineplot(data=flights, x='year', y='passengers', hue='month', palette='ch:.50')
plt.legend(bbox_to_anchor=(1.03, 1), loc=2) #legend 그래프 밖에 추가하기
```
