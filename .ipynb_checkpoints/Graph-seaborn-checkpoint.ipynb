#### Seaborn palette
Seaborn의 palette는 그래프에서 사용할 색상을 지정하는 역할을 합니다. 이를 통해 그래프에서 각 데이터 포인트에 대한 구분이 쉽게 되며, 시각화를 보다 효과적으로 할 수 있습니다.

palette는 일반적으로 color_palette() 함수를 통해 지정됩니다. 이 함수는 다양한 인자를 받아 Seaborn에서 사용할 색상 팔레트를 생성하며, 이를 palette 파라미터에 전달합니다.

palette에 전달할 수 있는 인자는 다음과 같습니다.

문자열: 미리 정의된 Seaborn 색상 팔레트 중 하나를 사용할 수 있습니다. 예를 들어 palette="Blues"와 같이 지정하면 파란색 계열의 색상 팔레트가 사용됩니다.

리스트: 색상 팔레트를 직접 지정할 수 있습니다. 리스트의 각 원소는 색상을 지정하는 문자열이어야 합니다. 예를 들어 palette=["red", "blue", "green"]와 같이 지정하면 빨간색, 파란색, 녹색의 색상 팔레트가 사용됩니다.

색상 맵: Matplotlib 라이브러리에서 제공하는 색상 맵 중 하나를 사용할 수 있습니다. 예를 들어 palette="viridis"와 같이 지정하면 Viridis 색상 맵이 사용됩니다.

딕셔너리: 그래프에서 사용할 데이터 카테고리와 색상을 지정할 수 있습니다. 예를 들어 palette={"cat": "red", "dog": "blue", "bird": "green"}와 같이 지정하면 각 카테고리에 대한 색상을 지정할 수 있습니다.

Seaborn에서 지원하는 색상 팔레트의 종류와 사용 방법 등에 대한 더 자세한 내용은 Seaborn 공식 문서를 참고

https://seaborn.pydata.org/tutorial/color_palettes.html

## Hist
히스토그램은 도수분포표를 그래프로 나타낸 것입니다. 
> ↔ 가로축
> - 계급: 변수의 구간, bin (or bucket)
>
> ↕ 세로축
> - 도수: 빈도수, frequency
>
> 전체 총량: n


x1은 평균은 100이고 표준편차는 15인 정규분포를 따릅니다.  

x2는 평균은 130이고 표준편차는 15인 정규분포를 따릅니다.  

도수를 50개의 구간으로 표시하며, 확률 밀도가 아닌 빈도로 표기합니다.

```python
import random 
```

```python
normal_l1 = []
normal_l2 = []
for i in range(10000):
    normal_l1.append(random.normalvariate(100,15))
    normal_l2.append(random.normalvariate(130,15))
```

```python
plt.hist(normal_l1)
plt.hist(normal_l2)
plt.show()
```

```python
fig = plt.figure()
ax1 =fig.add_subplot(2,1,1)
ax1.hist(normal_l1,bins=50)

ax2 = fig.add_subplot(2,1,2)
ax2.hist(normal_l2,bins=50)

plt.show()
```

data: 히스토그램을 그릴 데이터. numpy 배열이나 리스트 등의 시퀀스 형태여야 함.

bins: 히스토그램의 구간 개수. 기본값은 10.

range: 히스토그램 구간의 범위. 기본값은 데이터의 최솟값과 최댓값을 사용.

density: 정규화 여부를 지정하는 boolean 값. 기본값은 False.

weights: 각 데이터의 가중치를 지정하는 배열.

cumulative: 누적 히스토그램 여부를 지정하는 boolean 값. 기본값은 False.

color: 히스토그램 막대의 색상.

alpha: 히스토그램 막대의 투명도.

histtype: 히스토그램 막대 형태를 지정하는 문자열. 'bar', 'barstacked', 'step', 'stepfilled' 등의 값을 가질 수 있음.

align: 히스토그램 막대의 정렬 방식을 지정하는 문자열. 'left', 'mid', 'right' 값을 가질 수 있음.

orientation: 히스토그램 막대의 방향을 지정하는 문자열. 'horizontal', 'vertical' 값을 가질 수 있음.

rwidth: 히스토그램 막대의 너비를 지정하는 상대적인 값. 기본값은 0.9.

log: y축 스케일을 로그 스케일로 지정할지 여부를 지정하는 boolean 값. 기본값은 False.

```python
fig = plt.figure()
plt.title('Tow Frequency Distribusion')

ax1 =fig.add_subplot(2,1,1)
ax1.hist(normal_l1,bins=50,density=True,color='blue',alpha=0.5)
ax1.hist(normal_l2,bins=50,density=True,color='red',alpha=0.5)


ax2 = fig.add_subplot(2,1,2)
ax2.hist(normal_l1,bins=50,color='blue',alpha=0.5)
ax2.hist(normal_l2,bins=50,color='red',alpha=0.5)



plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 1, 1)
ax1.hist(normal_l1, bins=50, density=True, color='blue', alpha=0.5)
ax1.hist(normal_l1, bins=50, density=True, color='red', alpha=0.5)

plt.show()
```

```python
sns.histplot(df['total_bill'], label="total_bill")
sns.histplot(df['tip'], label="tip").legend()
```

**Seaborn histplot() Parameter**
- data: 히스토그램을 그릴 데이터. numpy 배열이나 pandas DataFrame이어야 함.

- bins: 히스토그램의 구간 개수. 기본값은 'auto'로 데이터 개수에 따라 자동으로 결정됨.

- range: 히스토그램 구간의 범위. 기본값은 데이터의 최솟값과 최댓값을 사용.

- density: 정규화 여부를 지정하는 boolean 값. 기본값은 False.

- weights: 각 데이터의 가중치를 지정하는 배열.

- cumulative: 누적 히스토그램 여부를 지정하는 boolean 값. 기본값은 False.

- hue: 데이터를 분할하는 기준이 되는 변수.

- multiple: hue에 따라 여러 개의 히스토그램을 그릴지 여부를 지정하는 boolean 값. 기본값은 False.

- stat: y축 값의 기준을 지정하는 문자열. 'count', 'frequency', 'density', 'probability' 값을 가질 수 있음. 기본값은 'count'.

- bins: 히스토그램의 구간 개수를 지정합니다.

- kde: 밀도 그래프를 추가할지 여부를 지정하는 boolean 값. 기본값은 False.

- kde_kws: 밀도 그래프의 세부적인 설정을 지정하는 딕셔너리.

- element: 히스토그램을 구성하는 요소를 지정하는 문자열. 'bars', 'step', 'poly' 등의 값을 가질 수 있음.

- fill: 히스토그램 막대의 채움 여부를 지정하는 boolean 값. 기본값은 True.

- color: 히스토그램 막대의 색상.

- alpha: 히스토그램 막대의 투명도.

- linewidth: 히스토그램 막대의 선 굵기.

- edgecolor: 히스토그램


**밀도 그래프: 연속된 확률분포를 나타냅니다.**
- 일반적으로는 kernels메서드를 섞어서 이 분포를 근사하는 식으로 그립니다.
- 이것은 좀 더 단순하고 우리에게 친숙한 정규분포(가우시안)로 나타낼 수 있습니다.
- 위 밀도 그래프는 KDE(Kernel Density Estimate) 커널 밀도 추정 그래프입니다.

[KDE 자세한 설명](https://darkpgmr.tistory.com/147#:~:text=Kernel%20Density%20Estimation%20(%EC%BB%A4%EB%84%90%20%EB%B0%80%EB%8F%84%20%EC%B6%94%EC%A0%95)%20%EB%B0%A9%EB%B2%95%EC%9D%80%20non%2D,%EC%9D%84%20%EA%B0%9C%EC%84%A0%ED%95%9C%20%EB%B0%A9%EB%B2%95%EC%9D%B4%EB%8B%A4.)
